# Estadística Descriptiva

Una de las etapas más importantes en el análisis y procesamiento de los datos es la comprensión de los mismos, ya que al entenderlos se pueden plantear diferentes técnicas, algoritmos y procedimientos que se pueden utilizar para la construcción de modelos.

En los siguientes fragmentos de código se procede rápidamente con la carga, preparación y análisis exploratorio de los datos correspondientes a <a href="https://archive.ics.uci.edu/ml/datasets/Computer%2BHardware" target="_blank">
Computer Hardware Data Set</a>; Para efectos de este conjunto de datos es necesario definirle las cabeceras con los nombres de las columnas, puesto que no vienen incluidas en la base de datos, por lo tanto se llena la variable headers y se le especifican al momento de definir la variable dataset.

```python
!pip install pandas
!pip install requests

import pandas as pd
import requests

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

import requests
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/cpu-performance/machine.data"
data_stream = requests.get(url).content

import io

headers = ['VENDOR_NAME', 'MODEL_NAME', 'MYCT', 'MMIN', 'MMAX', 'CACH', 'CHMIN', 'CHMAX', 'PRP', 'ERP']
dataset=pd.read_csv(io.StringIO(data_stream.decode('utf-8')), header=None, names = headers)

dataset
```

```python
dataset.info()
```

Como vemos al final de la ejecución del anterior fragmento de código la carga de los datos se realizo correctamente; adicionalmente y para efectos del análisis las variables independientes son **MYCT, MMIN, MMAX, CACH, CHMIN, CHMAX, PRP** y la variable dependiente o de decisión es **ERP** las cuales se especificaran a continuación.

**Nota.** Se definen de esta forma las variables de interés puesto que el proposito de estos datos esta enfocado en tratar de describir la variable **ERP.** Adicionalmente las variables **VENDOR_NAME y MODEL_NAME** no se tendran en cuenta puesto que a pesar que pueden ser utiles en el proceso nos tocaría realizar una serie de transformaciones para poder trabajar con ellas y para efectos de este ejercicio no se realizaran dichas transformaciones porque estan mas ligadas a la sección de preparción de los datos. Entonces aca estamos 

```python
x_columns =  ['MYCT', 'MMIN', 'MMAX', 'CACH', 'CHMIN', 'CHMAX', 'PRP']
target_column_name = 'ERP'
y_columns = [target_column_name]
```

Teniendo en cuenta que en condiciones normales en el ejercicio de analisis de datos, dividan los datos en dos partes.

1. Subconjunto para descubrimiento, tratamiento y entrenamiento,
2. Subconjunto para validación de los modelos.

Esta división la vamos a encontrar en la ejecución de todos los ejemplos puesto que hace parte del flujo de trabajo.

```python
# El primer parámetro nos va a servir para tomar X porcentaje de los datos por ejemplo 0.6 quiere decir el 60% de los datos
train_data=dataset.sample(frac=0.6)
# Lo que se hace aca es tomar los elementos del conjunto de datos que no se encuentran en el conjunto de entrenamiento
test_data=dataset.drop(train_data.index)
```

Una de las bondades y ventajas de Python, es que con comandos muy simples nos permite conocer rapidamente los datos, en este caso vamos a utilizar la funcion **describe()** ya que nos devuelve la información correspondiente a:

* Cantidad de observaciones,
* Promedio
* Desviación estandar
* Los valores mínimo y máximo
* La ubicación de los valores relacionados a los cuartiles.

Obteniendo en un primer vistazo las medidas de tendencia central y detectar el rango de los valores que puede tener cada variable.

```python
train_data.describe()
```

Teniendo en cuenta que la visualización de las medidas a lo mejor no son tan fáciles de leer, vamos a trasponer la presentación de las medidas para facilitar la presentación utilizando la función **transpose()**

```python
train_data.describe().transpose()
```

Ahora vamos a hacer una visualización general de los datos en la cual tendremos el diagrama de dispersión de los datos al cruzar las variables, y en la diagonal encontraremos el histograma asociado a la variable.

**Nota.** Se pueden hacer otros diagramas segun sea el interés del cientifico de datos, en este caso se toman estos gráficos para intentar intuir la posible implementación de algún algoritmo de machine learning.

```python
import seaborn as sns
import matplotlib.pyplot as plt
g = sns.pairplot(train_data)
plt.show()
```

A simple vista si revisamos a profundidad cada uno de los gráficos podemos destacar el diagrama asociado a las variables **PRP** y **ERP** puesto que se puede intuir rápidamente que hay una relación fuerte o proporcional entre dichas variables. 

```python
plt.plot(train_data['PRP'], train_data['ERP'], 'o')
plt.xlabel('PRP')
plt.ylabel('ERP')
```

Ahora para comprender un poco más los datos es pertinente tener la matriz de correlación de variables; ya que con ella se puede evidenciar a mayor detalle

```python
# Coeficientes de correlación
correlation_matrix = train_data.corr()
correlation_matrix
```

A lo mejor al tener a simple vista la matriz de correlación es difícil interpretarla, pero con el siguiente fragmento de código vamos a obtener una gráfica para entenderla más rápidamente mediante códigos de color. A mayor correlación, el color será mas claro y si hay correlación negativa vamos a encontrar colores mas oscuros, y si la correlación es cercana a cero el color va a ser más tirando al morado.

```python
import seaborn as sn

sn.heatmap(correlation_matrix, annot=True)
plt.show()
```

Como se puede observar en los pasos anteriores como en el que se encontro el coeficiente de correlación entre las variables, y en las graficas. Teniendo en cuenta que la variable ERP es la variable sobre la cual queremos enfocar el análisis podemos afirmar:
1. Dado que el coeficiente de correlación entre la variable PRP y ERP es 0.975190, entonces podemos afirmar que la variable ERP puede ser representada en terminos de un modelo de regresión lineal simple con la variable PRP.
2. Si quisieramos representar el modelo de regresion lineal multivariado, podriamos ademas de la variable PRP podriamos utilizar las variables MMIN y MMAX las cuales poseen altos coeficientes de correlación 0.835403	y 0.915207 respectivamente.
3. No obstante podriamos a modo de practica hacer la regresion lineal con la totalidad de atributos, tambien excluyendo la variable MYCT; dado que el coeficiente de correlación de dicha variable es muy bajo.
4. También podriamos usar algún algoritmo de clasificación o de agrupamiento, que le pueda dar un mejor tratamiento al tema de la dispersión de la información.
